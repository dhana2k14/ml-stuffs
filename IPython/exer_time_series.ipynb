```python
from pandas import DataFrame, Series, TimeGrouper
from matplotlib import pyplot
import pandas as pd
from pandas import concat
%matplotlib inline
```

```python
series = Series.from_csv('D://daily-minimum-temperatures-in-me.csv', sep = ',', header = 0)
one_yr = series['1990']
```

```python
print type(series)
```

```python
groups = one_yr.groupby(TimeGrouper('M'))
```

```python
months = concat([DataFrame(x[1].values) for x in groups], axis = 1)
```

```python
months.columns = range(1,13)
```

```python
values = DataFrame(series.values)
lags = 7
columns = [values]
for i in range(1,(lags + 1)):
    columns.append(values.shift(i))
dataframe = concat(columns, axis=1)
```
