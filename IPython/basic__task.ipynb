i. Create a simulated dataset of at least 1000 records, with ID column having sequence number ranging from 1 to 1000, additional columns with the constraints as follows.
0 0<= A,B,C,D,E,F <=100 and Class column with randomly populated integers of 0 and 1.
ii. Split the dataset into train / test sets having 80% and 20% records respectively and store them in a seperate csv files. 
iii. Plot the train and test set plots with different labels on both X and Y 

```python
import pandas as pd, numpy as np, random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
cl = ['0', '1']
df = pd.DataFrame({'ID': range(1,1001,1),
                  'A': np.random.randint(0, 100, size = 1000), 
                  'B': np.random.randint(0, 100, size = 1000),
                  'C': np.random.randint(0, 100, size = 1000),
                  'D': np.random.randint(0, 100, size = 1000),
                  'E': np.random.randint(0, 100, size = 1000),
                  'F': np.random.randint(0, 100, size = 1000),
                  'Class': np.random.choice(cl, 1000, p = [0.5, 0.5])})
```
